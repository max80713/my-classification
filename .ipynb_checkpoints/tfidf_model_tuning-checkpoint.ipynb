```python
import json
import operator
```

```python
tf_dict = json.load(open('tf_dict.json','r'))
idf_dict = json.load(open('idf_dict.json','r'))
tfidf_dict = json.load(open('tfidf_dict.json','r'))
train_data = json.load(open('train_data.json','r'))
test_data = json.load(open('test_data.json','r'))
```

```python
for event_class, keyword_tfidf_dict in tfidf_dict.items():
    tfidf_dict[event_class] = sorted(keyword_tfidf_dict.items(), key=operator.itemgetter(1), reverse=True)
```

```python
for event_class, keyword_tfidf_tuple in tfidf_dict.items():
    print(event_class)
    count = 0
    for keyword_tfidf_pair in keyword_tfidf_tuple:
        count += 1
        print(keyword_tfidf_pair)
        if count == 20:
            break
```
